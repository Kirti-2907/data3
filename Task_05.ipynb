# Task 05 : Question_1

```python
import matplotlib.pyplot as plt
import numpy as np

years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
sachin = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
kohli =  [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
```

```python
# Select 5 years
#This line defines which years you want to compare the runs.
selected_years = [1994, 1996, 2003, 2007, 2010]
indices = [years.index(y) for y in selected_years] #This line finds the positions (index numbers) of those selected years from the full years list.

# These three lines pick the runs of Sachin, Kohli, and Sehwag for only the 5 selected years.
sachin_selected = [sachin[i] for i in indices]
kohli_selected = [kohli[i] for i in indices]
sehwag_selected = [sehwag[i] for i in indices]

# Bar chart
x = np.arange(len(selected_years)) #This creates positions on the X-axis: [0, 1, 2, 3, 4]
bar_width = 0.25  #This sets the width of each bar in the bar chart

plt.figure(figsize=(10,6))  #This creates a new figure (chart canvas) that is 10 units wide and 6 units tall.
plt.bar(x, sachin_selected, width=bar_width, label='Sachin')  #Draws Sachin’s bars at positions [0, 1, 2, 3, 4]
plt.bar(x + bar_width, kohli_selected, width=bar_width, label='Kohli') #x + bar_width moves each bar to the right by 0.25.
plt.bar(x + 2*bar_width, sehwag_selected, width=bar_width, label='Sehwag') #Sehwag, shifted even more right so all three bars are side-by-side.

plt.xticks(x + bar_width, selected_years)  #The x + bar_width ensures the labels are centered below the middle bar group.
plt.xlabel("Year") 
plt.ylabel("Runs")
plt.title("Runs Comparison Over 5 Years")
plt.legend()
plt.show()
```

# question_01 : Horizontal Bar chart 

```python
# Total runs in debut 5 years (picked from above data)
sachin_total = sum([500, 700, 1100])      # 1990–1994 This line adds up Sachin’s runs from his first 5 years (1990, 1992, 1994) #500 + 700 + 1100 = 2300
kohli_total = sum([1500, 1800, 1900])     # 2003–2007 #1500 + 1800 + 1900 = 5200
sehwag_total = sum([1500, 1700])          # 1998–2002 (2 years) #Sehwag's total in debut 5 years — here, only 2 years of data are used (1998, 2000)

players = ['Sachin', 'Kohli', 'Sehwag'] #the list of player names to display on the Y-axis of the chart.
totals = [sachin_total, kohli_total, sehwag_total] ##This list contains the total runs each player scored in their first 5 years.

plt.figure(figsize=(8,5)) 
plt.barh(players, totals, color='#E08D8D')
plt.xlabel("Runs")
plt.title("Total Runs in Debut 5 Years")
plt.show()
```

```python
# Market share data
labels = ['Android', 'iOS', 'Others']
sizes = [72, 25, 3]

# Pie Chart
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Mobile OS Market Share")
plt.show()

# Horizontal Bar Chart
plt.figure(figsize=(8, 4))
plt.barh(labels, sizes, color=['green', 'orange', 'skyblue'])
plt.xlabel("Market Share (%)")
plt.title("Mobile OS Market Share (Bar View)")
plt.tight_layout()
plt.show()
```

```python
#4
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 300)

y_sin = np.sin(x)
y_cos = np.cos(x)
y_exp = np.exp(0.5 * x)  # 0.5 factor keeps exponential from growing too large

plt.plot(x, y_sin, color='red', linestyle='-', label='sin(x)')
plt.plot(x, y_cos, color='blue', linestyle='--', label='cos(x)')
plt.plot(x, y_exp, color='green', linestyle='-.', label='exp(0.5x)')

plt.xlabel('x')
plt.ylabel('y')
plt.title('sin(x), cos(x) and exp(0.5x) on the same axes')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#6
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 6)
y = np.array([2, 5, 7, 1, 6])

fig, axs = plt.subplots(2, 2, figsize=(8, 6))

# 1:--Line plot
axs[0, 0].plot(x, y, color='#ff6666')
axs[0, 0].set_title("Line Plot")

# 2:--Bar chart
axs[0, 1].bar(x, y, color='#df80ff')
axs[0, 1].set_title("Bar Chart")

# 3:--Scatter plot
axs[1, 0].scatter(x, y, color='#9999ff')
axs[1, 0].set_title("Scatter Plot")

# 4:--Histogram
data = np.random.randn(100)
axs[1, 1].hist(data, bins=10, color='#d9ff66', edgecolor='black')
axs[1, 1].set_title("Histogram")

fig.suptitle("2x2 Subplots Example", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
#9
import matplotlib.pyplot as plt

companies = ['CyberLink', 'NovaSoft', 'QuantumEdge', 'BlueOrbit']
market_share = [38, 27, 21, 14]

largest_index = market_share.index(max(market_share))

explode = [0.1 if i == largest_index else 0 for i in range(len(market_share))]

colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']

plt.figure(figsize=(7,7))
plt.pie(market_share,
        labels=companies,
        explode=explode,
        autopct='%1.1f%%',
        startangle=90,
        colors=colors,
        shadow=True)
plt.title("Market Share Distribution(2025)", fontsize=14)
plt.show()
```
